```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Get URLs from PudMed search results

## Download article details from each URL

```python
def read_article_from_pubmed(url):
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    data = {}
    
    element_title = soup.select_one('.rprt_all h1')
    element_abstract = soup.select_one('.abstr')  
    element_citation = soup.select_one('.cit')
    
    data['title'] = "Not found" if element_title is None else element_title.text
    data['abstract'] = "Not found" if element_abstract is None else element_abstract.text
    data['url'] = url
    data['citation'] = "Not found" if element_citation is None else element_citation.text
    data['authors'] = ', '.join([item.text for item in soup.select('.auths a')])
    return data
    
article_datas = []
index = 1
for url in article_urls:
    print(f'{index}. Downloading details from {url}')
    article_datas.append(read_article_from_pubmed(url))
    index = index + 1
```

## Export the downloaded data to CSV/ HTML files

```python
df = pd.DataFrame(article_datas)
df.to_csv('output/pubmed.csv', index = False)
df.head()
```
